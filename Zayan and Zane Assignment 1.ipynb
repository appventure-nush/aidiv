## Dataset Preprocessing

Importing of all the important libraries

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA
# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import RocCurveDisplay
```

Loading of mushroom dataset and renaming the columns

```python
mushroom_dataset = pd.read_csv("mushroom_data/agaricus-lepiota.data", header=None)
mushroom_dataset.head()
```

```python
columns = ["edibility", "cap-shape", "cap-surface", "cap-color", "bruises", "odor", "gill-attatchment",
          "gill-spacing", "gill-size", "gill-color", "stalk-shape", "stalk-root", "stalk-surface-above-ring",
          "stalk-surface-below-ring", "stalk-color-above-ring", "stalk-color-below-ring", "veil-type",
          "veil-color", "ring-number", "ring-type", "spore-print-color", "population", "habitat"]
mushroom_dataset.columns = columns
mushroom_dataset.head()
```

cleaning unknown values

```python
mushroom_dataset = mushroom_dataset.replace("?", np.nan)
mushroom_dataset.info()
```

Removing stalk-root because it has a lot of null data and veil-type has only 1 class of data.

```python
mushroom_dataset = mushroom_dataset.drop(["stalk-root", "veil-type"], axis = 1)
mushroom_dataset.info()
```

Convert the data to numbers

```python
lookup_dict = dict()

for i in mushroom_dataset.columns:
    temp_dict = dict()
    i_unique = mushroom_dataset[i].unique()
    for j in range(len(i_unique)):
        mushroom_dataset[i] = mushroom_dataset[i].replace(i_unique[j], j)
        temp_dict[j] = i_unique[j]
    lookup_dict[i] = temp_dict
    
mushroom_dataset.head()
```

Lookup dictionary

```python
for key, value in lookup_dict.items():
  print(key + ": " + str(value))
```

Correlation matrix between all the variables

```python
corr_matrix = mushroom_dataset.corr().round(2)
plt.subplots(figsize=(20, 10))
sns.heatmap(data=corr_matrix,annot=True,cmap="Blues")
plt.show()
```

## Principal Component Analysis

Performing PCA on the data since their are a large amount of variables to prevent overfitting

```python
x = mushroom_dataset.loc[:, mushroom_dataset.columns[1:]].values
x = StandardScaler().fit_transform(x)
pca = PCA().fit(x)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative explained variance") 
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

We will use 8 principal components as the 8 principal components account for 80% of the variance

```python
principal_components = pca.transform(x)
principal_component_df = pd.DataFrame(principal_components)
```

Renaming and cleaning the dataframe

```python
df_cleaned = pd.concat([principal_component_df.iloc[:, :8], mushroom_dataset[["edibility"]]], axis = 1)
df_cleaned.columns = ["pc" + str(i) for i in range(1,9)] + ["edibility"]
df_cleaned.head()
```

Scatterplot graph

```python
colors=['red','green']
for color, i,target in zip(colors,[0,1],list(lookup_dict['edibility'].values())):
    #print(color,i, target)
    plt.scatter(df_cleaned[df_cleaned['edibility']==i]['pc1'],
                df_cleaned[df_cleaned['edibility']==i]['pc2'], color=color,label=target)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(loc='best',shadow=False,scatterpoints=1)

plt.show()
```

Getting the training data

##  Train-Test Split

We split the data into training and testing data. 80% of the data is used for training and 20% is used for testing

```python
X_data = df_cleaned.iloc[:, :8].values
y_data = df_cleaned['edibility']
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=5)
```

## Decision Tree Classifier

The Decision Tree Classifier is a classification model which makes it appropriate for this task. It works by selecting the node with the most information given and recursing downwards allowing us to create a tree of decisions to classify the test data.

```python
dt_clf=DecisionTreeClassifier(random_state=0)
start_time = time.time()
dt_clf.fit(X_train,y_train)
end_time = time.time()
print("Time taken: {:.3f}s".format(end_time - start_time))
print("Validation Accuracy: {:.1f}%".format(dt_clf.score(X_test,y_test) * 100)) 
```

## Random Forest Classifier

The Ransdom Tree Classifier is a classificaiton model which makes it appropriate for this task. It is basically a decision tree but the decision is randomized and has better performance than the deterministic version. 

```python
clf=RandomForestClassifier(n_jobs=2, random_state=0)
start_time = time.time()
clf.fit(X_train,y_train)
end_time = time.time()
print("Time taken: {:.3f}s".format(end_time - start_time))
print("Validation Accuracy: {:.1f}%".format(clf.score(X_test,y_test) * 100)) 
```

## Support Vector Machine

Another one of the algorithms we picked is the support vector machine or SVM. SVM is a classification algorithm  which  The SVM works by finding a function seperating the group of data points in a n-dimensional space by their class.

```python
svm=SVC()
start_time = time.time()
svm.fit(X_train,y_train)
end_time = time.time()
print("Time taken: {:.3f}s".format(end_time - start_time))
print("Validation Accuracy: {:.1f}%".format(svm.score(X_test,y_test) * 100)) 
```

## Logistic Regression

Logisitc regression is a classification method for binary classifcation problems. It uses the data to fit a non-linear function known as the logistic value which outputs a value between 0 or 1 which indicates the ouput of the model.

```python
logreg=LogisticRegression()
start_time = time.time()
logreg.fit(X_train,y_train)
end_time = time.time()
print("Time taken: {:.3f}s".format(end_time - start_time))
print("Validation Accuracy: {:.1f}%".format(logreg.score(X_test,y_test) * 100)) 
```

## Artificial Neural Network

Lastly, we also use a neural network just for fun and to test out how it compares against the other machine learning algorithms. The neural network has 2 hidden layers with 10 and 6 perceptron respectively.

```python
ann=MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(10, 6), random_state=1)
start_time = time.time()
ann.fit(X_train,y_train)
end_time = time.time()
print("Time taken: {:.3f}s".format(end_time - start_time))
print("Validation Accuracy: {:.1f}%".format(ann.score(X_test,y_test) * 100)) 
```

For our results analysis we plotted the confusion matrix and we printed the classification report. The confusion matrix shows True Positives in the top left corner, the False Positives in the top right corner, the True Negatives in the bottom right corner and the False Negatives in the bottom left corner. The classification report shows precision, recall, f1-score and support. We also showed the ROC curve. This is a receiver operating characteristic(ROC) curve which signify the true positive vs false positive rate at different classification thresholds. Classification thresholds are basically what is the boundary between Edible and Not edible in our case. The area under curve(AUC) provides an aggregate across all possible classification thresholds. The more accurate the model the larger the AUC value.

```python
disp=RocCurveDisplay.from_estimator(dt_clf, X_test, y_test)
```
